```python
import numpy as np
import pandas as pd
```

```python
num_of_models = 5
num_of_metrics = 6
```

```python
V = np.absolute( np.random.normal(0, 1, size=(num_of_models, num_of_metrics)) )


weights = np.random.uniform(0,1,num_of_metrics)
weights = weights/weights.sum()
weights
```



> **Дистрибутивний метод**



```python
r_koefs_denom =  np.sum(V, axis=0)
r_koefs_denom_mat = np.repeat(r_koefs_denom, num_of_models).reshape(num_of_metrics, num_of_models)
r_koefs_denom_mat = np.transpose(r_koefs_denom_mat)

r_koef_mat = np.divide(V, r_koefs_denom_mat)

weights_mat = np.repeat(weights, num_of_models).reshape(num_of_models, num_of_metrics)

distributed_rating_mat = np.multiply(weights_mat, r_koef_mat)

distributed_rating = distributed_rating_mat.sum(axis=1)
distributed_rating
```

```python
r_koefs_denom =  np.max(V, axis=0)
r_koefs_denom_mat = np.repeat(r_koefs_denom, num_of_models).reshape(num_of_metrics, num_of_models)
r_koefs_denom_mat = np.transpose(r_koefs_denom_mat)

r_koef_mat = np.divide(V, r_koefs_denom_mat)

weights_mat = np.repeat(weights, num_of_models).reshape(num_of_models, num_of_metrics)

distributed_rating_mat = np.multiply(weights_mat, r_koef_mat)

modified_distributed_rating = distributed_rating_mat.sum(axis=1)
modified_distributed_rating
```

```python
weights_mat = np.repeat(weights, num_of_models).reshape(num_of_models, num_of_metrics)

power_mat = np.power(V, weights_mat)

power_rating = power_mat.prod(axis=1)
power_rating
```

```python
weights_mat = np.repeat(weights, num_of_models).reshape(num_of_models, num_of_metrics)
minimum_rating_mat = np.multiply(V, weights_mat)
minimum_rating = minimum_rating_mat.min(axis=1)
minimum_rating
```

```python
weights_agg = np.array([distributed_rating, modified_distributed_rating, power_rating, minimum_rating])
distributed_rating = distributed_rating/np.sum(distributed_rating)
modified_distributed_rating = modified_distributed_rating/np.sum(modified_distributed_rating)
power_rating = power_rating/np.sum(power_rating)
minimum_rating = minimum_rating/np.sum(minimum_rating)
print("distributed_rating: ")
print(distributed_rating)
print("modified_distributed_rating: ")
print(modified_distributed_rating)
print("power_rating: ")
print(power_rating)
print("minimum_rating: ")
print(minimum_rating)
```
